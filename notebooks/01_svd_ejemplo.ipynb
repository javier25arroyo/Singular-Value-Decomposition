# Ejemplo rápido de SVD
Este notebook muestra un ejemplo mínimo usando las funciones del paquete `proyecto_svd`.

```python
import numpy as np
from proyecto_svd.main import svd, reconstruct

A = np.array([[3, 1, 1],
              [-1, 3, 1]], dtype=float)
U, s, VT = svd(A, full_matrices=False)
A_rec = reconstruct(U, s, VT)
U, s, VT, A_rec
```
